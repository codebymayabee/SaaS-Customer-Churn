Which subscription plan has the highest churn rate?
churn percentage for each plan

```python
plan_churn = pd.crosstab(df["plan_type"], df["churn"])

print(plan_churn)

plan_churn_percent = pd.crosstab(
    df["plan_type"],
    df["churn"],
    normalize="index"
) * 100

plan_churn_percent
```

Premium plan customers have the highest churn rate (58.05%), followed closely by Basic (57.85%). Standard plan customers have the lowest churn rate (56.06%). Although the differences are relatively small, the Premium plan shows slightly higher customer attrition, suggesting the company should evaluate whether Premium users' expectations are being met.

Do customers who use the product less churn more?

```python
df.groupby("churn")["avg_weekly_usage_hours"].mean()
```

Customers who remained subscribed use the platform more frequently than customers who churned. Lower product engagement appears to be associated with a higher likelihood of churn, indicating that inactive users are at greater risk of leaving.

Do payment failures affect churn?


```python
df.groupby("churn")["payment_failures"].mean()
```

Customers who churned experienced more payment failures on average than retained customers. Payment issues may contribute to customer dissatisfaction and increase the likelihood of subscription cancellation.

Do customers with more support tickets churn?

```python
df.groupby("churn")["support_tickets"].mean()
```

Customers who churned raised more support tickets than those who remained. This suggests that unresolved issues or poor customer experience may contribute to customer churn.




Does customer tenure matter?

```python
df.groupby("churn")["tenure_months"].mean()

```

The average tenure is nearly identical for both churned and retained customers. This indicates that customer tenure alone does not appear to significantly influence churn in this dataset.



Do inactive users churn?

```python
df.groupby("churn")["last_login_days_ago"].mean()
```

Customers who churned had not logged into the platform for a longer period than retained customers. Reduced login activity appears to be an early warning sign of potential churn.


• Improve customer engagement through personalized recommendations and onboarding.

• Monitor inactive customers and launch re-engagement campaigns before they churn.

• Reduce payment failures by offering multiple payment options and automated reminders.

• Enhance customer support to resolve issues quickly and improve customer satisfaction.

• Review Premium subscription features and pricing to increase customer retention.

• Use churn prediction models in the future to identify at-risk customers proactively.
